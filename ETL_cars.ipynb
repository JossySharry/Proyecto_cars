```python
# Importamos las librerias a utilizar
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Cargamos el dataset
df = pd.read_csv("../Analisis de mercado/ML_cars.csv")
```

```python
df.head()
```

```python
df.info()
# Se observa que el dataframe cuenta con 26 columnas y 205 filas, donde no se tiene valores nulos
```

```python
df.shape
```

Se ha observado que la columna 'CarName', contiene la marca del vehículo y el modelo por lo que se procede a separar esta columna en dos 

```python
# Separamos la columna "CarName" en dos Marca y Modelo
df[['Marca', 'Modelo']] = df['CarName'].str.split(' ', n=1, expand=True)
```

```python
df.head()
```

```python
# Eliminamos la columna 'CarName' del dataframe
df = df.drop('CarName', axis=1)
```

```python
# Observamos lo valores únicos de la columna Marca 
Marca_unica = df['Marca'].unique()

# Se imprime los vlaores unicos de la Marca
print("Marca unica:")
print(Marca_unica)
```

Al imprimir las Marca, se puede observar que existe error de digitación por lo que se deben uniformizar los datos de la columna Marca

```python
# Se procede a estandarizar los nombres de las Marcas
Est_Marca = {'alfa-romero': 'Alfa Romero',
 'audi': 'Audi',
 'bmw':'BMW',
 'chevrolet': 'Chevrolet',
 'dodge': 'Dodge',
 'honda': 'Honda',
 'isuzu': 'Isuzu',
 'jaguar': 'Jaguar',
 'maxda': 'Mazda',
 'mazda':'Mazda',
 'buick': 'Buick',
 'mercury':'Mercury',
 'mitsubishi': 'Mitsubishi',
 'nissan':'Nissan',
 'peugeot':'Peugeot',
 'plymouth':'Plymouth',
 'porsche': 'Porsche',
 'porcshce': 'Porsche',
 'renault': 'Renault',
 'saab': 'Saab',
 'subaru':'Subaru', 
 'toyota':'Toyota',
 'toyouta': 'Toyota',
 'vokswagen': 'Volkswagen',
 'volkswagen': 'Volkswagen',
  'vw': 'Volkswagen',
  'volvo': 'Volvo'}
```

```python
df['Marca']=df['Marca'].replace(Est_Marca)
```

```python
# Observamos lo valores únicos de la columna Marca 
Marca_unica = df['Marca'].unique()

# Se imprime los vlaores unicos de la Marca
print("Marca unica:")
print(Marca_unica)
```

```python
# Observamos lo valores únicos de la columna Modelo 
Modelo_unico = df['Modelo'].unique()

# Se imprime los vlaores unicos de la Marca
print("Modelo unico:")
print(Modelo_unico)
```

```python
df.head()
```

```python
# Get the "car_ID" index
car_id_index = df.columns.get_loc("car_ID")

# Reordenar las columnas
new_columns_order = list(df.columns[:car_id_index]) + ['Marca', 'Modelo'] + list(df.columns[car_id_index:-2])

# Create new dataframe with changes
df_reordered = df[new_columns_order]

df = df_reordered
df = df.drop('car_ID', axis=1)
```

```python
df.head()
```

```python
# Obervamos la relacion existente entre la Marca y el Modelo
marca_modelo_relacion = df.groupby('Marca')['Modelo'].unique().reset_index()

# Se muestra los resultados 
print(marca_modelo_relacion)
```
